```python
import pandas as pd
dados3 = pd.read_csv('C:/Users/Andre/Documents/DidaticaTech/adesivo_teste.csv')
dados4 = pd.read_csv('C:/Users/Andre/Documents/DidaticaTech/adesivo_treino.csv')
pd.set_option('display.max_columns',42)
```

```python
dados4.info()
```

```python
from sklearn.preprocessing import LabelEncoder
pi = LabelEncoder()
dados4['date'] = pi.fit_transform(dados4['date'])
dados4['country'] = pi.fit_transform(dados4['country'])
dados4['store'] = pi.fit_transform(dados4['store'])
dados4['product'] = pi.fit_transform(dados4['product'])
```

```python
c = dados4['num_sold'].fillna(dados4['num_sold'].mean(), inplace = True)
c
```

```python
dados4.head(100)
```

```python
from sklearn.linear_model import LinearRegression
x = dados4.drop('num_sold', axis=1)
y = dados4['num_sold']

from sklearn.model_selection import train_test_split as tts
x_treino, x_teste, y_treino, y_teste = tts(x,y, test_size=0.3, random_state=10)
modelo = LinearRegression()
modelo.fit(x_treino, y_treino)
resultado = modelo.score(x_teste, y_teste)
print(resultado)
```
